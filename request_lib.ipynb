```python
import requests

# Example: Fetching NASA's Astronomy Picture of the Day
url = "https://api.nasa.gov/planetary/apod"
params = {"api_key": "DEMO_KEY"}  # free demo key

response = requests.get(url, params=params)
data = response.json()

print("Title:", data["title"])
print("URL:", data["url"])
```
