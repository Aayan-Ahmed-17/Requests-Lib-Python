```python
import requests
```

```python
# Example: Fetching NASA's Astronomy Picture of the Day
url = "https://api.nasa.gov/planetary/apod"
params = {"api_key": "DEMO_KEY"}  # free demo key

response = requests.get(url, params=params)
data = response.json()

print("Title:", data["title"])
print("URL:", data["url"])
```

**Get With Query Parameter**

```python
url = "https://pokeapi.co/api/v2/pokemon"
params = {"limit": 3, "offset": 0}    # fetch first 3 Pokémon

response = requests.get(url, params=params)
data = response.json()

# it holds all the fetched results
# data['results']

for pokemon in data['results']:
  print("Name:", pokemon["name"], "| URL:", pokemon["url"])
```

**POST Data using Request Lib**

```python
url = "https://httpbin.org/post"
payload = {"species": "Bald Eagle", "location": "Yellowstone"}

response = requests.post(url, json=payload)
print("Server response:", response.json())
```

**Handling Headers**

```python
url = "https://httpbin.org/headers"
headers = {"User-Agent": "MyNatureApp/1.0"}

response = requests.get(url, headers=headers)
print("Server response:", response.json())
```

**Timeouts and Errors**

```python
try:
  response = requests.get("https://httpbin.org/delay/5", timeout=2)
  print(response.json())
except requests.exceptions.Timeout:
  print("The request timed out")
```

**Working with Sessions**

```python
session = requests.Session()
session.headers.update({"User-Agent": "SoccerStatsBot/1.0"})

for team in ["Arsenal", "Barcelona", "Boca Juniors"]:
  r = session.get("https://httpbin.org/get", params = {"team": team})
  print(r.json()['args'])
```
